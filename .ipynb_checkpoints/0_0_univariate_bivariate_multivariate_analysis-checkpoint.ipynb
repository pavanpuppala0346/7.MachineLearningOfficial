```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/pavanpuppala0346/data/main/iris.csv")
df.head()
```

#### univariate

```python
df_setosa = df[df['species']=='setosa']
df_virginca = df[df['species']=='virginca']
df_versicolor = df[df['species']=='versicolor']
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginca['sepal_length'], np.zeros_like(df_virginca['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('sepal length')
plt.show()
```

#### Bivariate Analysis

```python
import seaborn as sns

sns.FacetGrid(df, hue='species', height=5).map(plt.scatter,'sepal_length','sepal_width').add_legend()
plt.show()
```

#### multi variate Analysis

```python
import seaborn as sns

sns.pairplot(df, hue='species', height=4)
```
