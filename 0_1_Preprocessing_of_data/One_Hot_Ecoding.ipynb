- from 6 categorical columns we end up with 117 columns
- if the columns increases it will impacts on accuracy so now we have to fix this
- To overcome this one hot encoding to the 10 most frequently labels of the variable. this means that they would make one binary variable for each of 10 most frequently labels only. this is equivalent to grouping all other variables under a new category, that in this case we will dropped. thus the 10 new dummy variables inidicate if one of the 10 most frequent labels is present (1) or (0) for a particular observation

```python
# find the top 10 most frquent categories for the variable X2

df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# top10 = [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
top10 = df.X2.value_counts().sort_values(ascending=False).head(10).index.to_list()
top10
```

```python
# we make 10 binary variables
import numpy as np

for label in top10:
    df[label] = np.where(df["X2"]==label,1,0)
```

```python
df[["X2"]+top10].head(20)
```

```python
def one_hot_encoding(df, variable, top_x_labels):
    for label in top_x_labels:
        df[variable+"-"+label] = np.where(df[variable]==label, 1, 0)
```

```python
df = pd.read_csv("mercedes_benz.csv", usecols=["X1","X2","X3","X4","X5","X6"])
df.head()
```

```python
one_hot_encoding(df, "X2", top10)
```
